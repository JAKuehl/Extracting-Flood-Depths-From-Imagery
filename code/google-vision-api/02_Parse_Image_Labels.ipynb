```python
# This function makes a new dataframe for rows that have 'truck' in one of the label columns
truck = [] 
for index, row in photos.iterrows():
    
    if row['label1'] == 'truck':
        truck.append(row)
    elif row['label2'] == 'truck':
        truck.append(row)
    elif row['label3'] == 'truck':
        truck.append(row)
    elif row['label4'] == 'truck':
        truck.append(row)
    elif row['label5'] == 'truck':
        truck.append(row)
    else:
        pass
trucks = pd.DataFrame(truck)
trucks
```
